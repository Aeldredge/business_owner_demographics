```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data12 = pd.read_csv('~/Documents/dsi/capstone/capstone1/data/2012/2012_data.csv')
data12.head()
```

```python
df12 = data12.groupby(['RACE_GROUP_TTL', 'ETH_GROUP_TTL','SEX_TTL']).agg({'FIRMALL':'sum'})
df12.rename(columns={'FIRMALL':'count'},inplace=True)
```

```python
df12.drop(index='Meaning of Race code',inplace=True)
```

```python
df12['count'] = pd.to_numeric(df12['count'])
```

Hypothesis Test on certain factors


```python
df12['percent'] = 100 * (df12['count'] / df12.iloc[0,0])
```

```python
percent12 = df12.sort_values('percent',ascending=False)
percent12
```

```python
consolidate12 = percent12[1:].drop(index='All firms',level=0)
consolidate12 = consolidate12[1:].drop(index='All firms',level=1)
consolidate12
```

```python
percent_race = consolidate12.groupby(level=0).agg({'percent':'sum'})
percent_race.sort_values('percent',ascending=False,inplace=True)
percent_race.reset_index(inplace=True)
percent_race.rename(columns={'RACE_GROUP_TTL':'race'},inplace=True)
percent_race
```

```python
zoomed_percent_race = percent_race[percent_race['race'] != 'White']
```

```python
x = np.arange(len(percent_race['race']))
x2 = np.arange(len(zoomed_percent_race['race']))

fig, ax = plt.subplots(1,2,figsize=(15,4))

ax[0].bar(x, percent_race['percent'])
ax[0].set_xticks(x)
ax[0].set_xticklabels(percent_race['race']);

plt.setp(ax[0].get_xticklabels(), rotation=45);

ax[1].bar(x2, zoomed_percent_race['percent'])
ax[1].set_xticks(x2)
ax[1].set_xticklabels(zoomed_percent_race['race']);

plt.setp(ax[1].get_xticklabels(), rotation=45);

ax[0].set_title("2012 Overall");
ax[1].set_title("2012 Zoomed");
```
